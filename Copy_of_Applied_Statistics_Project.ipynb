```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

```

```python
df = pd.read_csv('/content/Superstore Marketing Data - Sheet1.csv')
```

##Task 1 - Basic CleanUp

- **Clean and preprocess the dataset (handling missing values, data types, etc.).**

- **Analyze the distribution of customer demographics (age, education, marital status) using descriptive statistics and visualizations.**



**Deliverables**:

- **Cleaned and Preprocessed Dataset**:

  Provide a detailed report on the steps taken to handle missing values, including imputation methods used if applicable.

- **Summary of Basic Statistics**:

  Present calculated statistics such as mean, median, variance, and standard deviation for each relevant numerical variable.
  Include a concise table or summary showcasing these measures for easy reference.

```python
# Handling Missing Values
# Example: Replace missing values in the 'Year_Birth' column with the median birth year.
median_year_birth = df['Year_Birth'].median()
df['Year_Birth'].fillna(median_year_birth, inplace=True)
```

```python
# Data Type Consistency
# Ensure 'Education' and 'Marital_Status' columns are categorical
df['Education'] = df['Education'].astype('category')
df['Marital_Status'] = df['Marital_Status'].astype('category')
```

##Task 2 - Descriptive Statistics üìä

- **Calculate measures of central tendency (mean, median, mode) and measures of dispersion (variance, standard deviation) for key variables. Identify and handle outliers if necessary.**


**Deliverables**:

- **Descriptive statistics that reveal the central tendencies, variations, and potential outliers in the dataset.**:

  

```python
# Descriptive Statistics
# Calculate mean, median, variance, and standard deviation for 'Year_Birth'
year_birth_mean = df['Year_Birth'].mean()
year_birth_median = df['Year_Birth'].median()
year_birth_variance = df['Year_Birth'].var()
year_birth_std = df['Year_Birth'].std()


# Frequency distribution for 'Education'
education_counts = df['Education'].value_counts()

# Frequency distribution for 'Marital_Status'
marital_status_counts = df['Marital_Status'].value_counts()

# Visualizations
# Create a histogram for 'Year_Birth'
plt.hist(df['Year_Birth'], bins=20, color='skyblue')
plt.xlabel('Year of Birth')
plt.ylabel('Frequency')
plt.title('Year of Birth Distribution')
plt.show()

# Create a bar chart for 'Education'
education_counts.plot(kind='bar', color='green')
plt.xlabel('Education Level')
plt.ylabel('Count')
plt.title('Education Distribution')
plt.xticks(rotation=45)
plt.show()

# Create a pie chart for 'Marital_Status'
marital_status_counts.plot(kind='pie', autopct='%1.1f%%', colors=['gold', 'lightcoral', 'lightskyblue'])
plt.title('Marital Status Distribution')
plt.ylabel('')
plt.show()

# Print the summary statistics
print(f"Mean Year of Birth: {year_birth_mean}")
print(f"Median Year of Birth: {year_birth_median}")
print(f"Variance of Year of Birth: {year_birth_variance}")
print(f"Standard Deviation of Year of Birth: {year_birth_std}")
print("\nEducation Level Counts:")
print(education_counts)
print("\nMarital Status Counts:")
print(marital_status_counts)
```

## Task 3 - Probability Distributions üé≤

- **Identify variables that could follow specific probability distributions (e.g., Binomial, Normal). Calculate probabilities and expected values based on these distributions.**



**Deliverables**:

- **Determination of suitable probability distributions for relevant variables and corresponding calculated probabilities and expected values.**:

  

```python
# Calculate the probability of a positive response
positive_response_probability = (df['Response'] == 1).mean()
print(f"Probability of a Positive Response: {positive_response_probability:.2%}")


# Calculate mean and standard deviation for income
income_mean = df['Income'].mean()
income_std = df['Income'].std()
print(f"Mean Income: ${income_mean:.2f}")
print(f"Standard Deviation of Income: ${income_std:.2f}")
```

## Task 4: Insights and Customer Segmentation üìà

- **Explore relationships between customer characteristics and spending habits. Segment customers based on their behaviors and characteristics.**

**Deliverables**:

- **Key insights regarding relationships between variables and distinct customer segments based on behaviors.**

  

```python
# Explore the relationship between 'Education' and spending on 'MntWines'
education_groups = df.groupby('Education')['MntWines'].mean().sort_values(ascending=False)
print("Average Spending on Wines by Education Level:")
print(education_groups)

# Segment customers based on 'Education' and 'MntWines' spending
plt.figure(figsize=(10, 6))
plt.bar(education_groups.index, education_groups.values, color='skyblue')
plt.xlabel('Education Level')
plt.ylabel('Average Spending on Wines')
plt.title('Segmentation by Education and Wine Spending')
plt.xticks(rotation=45)
plt.show()
```
